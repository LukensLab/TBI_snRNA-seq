```python
from itertools import product
from scipy.stats import mannwhitneyu

def cliffs_delta(x, y):
    nx, ny = len(x), len(y)
    more = sum(1 for a, b in product(x, y) if a > b)
    less = sum(1 for a, b in product(x, y) if a < b)
    return (more - less) / (nx * ny)

# Ensure group labels are strings
adata.obs['group'] = adata.obs['group'].astype(str)
score_key = 'microglia_reactivity_index'

### 1. TBI-GFP vs Sham-GFP (expecting TBI-GFP > Sham-GFP)
g1_vals = adata.obs.loc[adata.obs['group'] == 'TBI-GFP', score_key].dropna().values
g2_vals = adata.obs.loc[adata.obs['group'] == 'Sham-GFP', score_key].dropna().values

stat1, p1 = mannwhitneyu(g1_vals, g2_vals, alternative='greater')
delta1 = cliffs_delta(g1_vals, g2_vals)

print("TBI-GFP vs Sham-GFP")
print(f"One-sided Mann–Whitney U p = {p1:.4f}")
print(f"Cliff's Delta = {delta1:.3f}\n")

### 2. TBI-GFP vs TBI-VEGFC (expecting TBI-GFP > TBI-VEGFC — rescue)
g1_vals = adata.obs.loc[adata.obs['group'] == 'TBI-GFP', score_key].dropna().values
g2_vals = adata.obs.loc[adata.obs['group'] == 'TBI-VEGFC', score_key].dropna().values

stat2, p2 = mannwhitneyu(g1_vals, g2_vals, alternative='greater')
delta2 = cliffs_delta(g1_vals, g2_vals)

print("TBI-GFP vs TBI-VEGFC")
print(f"One-sided Mann–Whitney U p = {p2:.4f}")
print(f"Cliff's Delta = {delta2:.3f}\n")

### 3. Sham-GFP vs TBI-VEGFC (expecting TBI-VEGFC > Sham-GFP)
g1_vals = adata.obs.loc[adata.obs['group'] == 'TBI-VEGFC', score_key].dropna().values
g2_vals = adata.obs.loc[adata.obs['group'] == 'Sham-VEGFC', score_key].dropna().values

stat3, p3 = mannwhitneyu(g1_vals, g2_vals, alternative='greater')
delta3 = cliffs_delta(g1_vals, g2_vals)

print("TBI-VEGFC vs Sham-VEGFC")
print(f"One-sided Mann–Whitney U p = {p3:.4f}")
print(f"Cliff's Delta = {delta3:.3f}")
```
